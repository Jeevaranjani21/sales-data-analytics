# Sales Data Analysis
This notebook analyzes the sales.csv dataset using Python.

## Steps Included:
- Load data using pandas
- Show summary statistics
- Find top-performing region
- Total sales
- Simple bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales.csv')
df
```

```python
# Summary statistics
df.describe()
```

```python
# Total sales
df['sales'].sum()
```

```python
# Sales by region
df.groupby('region')['sales'].sum()
```

```python
df.groupby('region')['sales'].sum().plot(kind='bar', figsize=(6,4))
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales by Region')
plt.show()
```
